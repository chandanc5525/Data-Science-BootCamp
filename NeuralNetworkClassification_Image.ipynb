```python
model = Sequential()

model.add(Dense(units = 1))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=300, batch_size=32)
```

```python
model = Sequential()

model.add(Dense(units = 1))
model.add(Dense(units = 1))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=300, batch_size=32)
```

```python
model.evaluate(X,y)
```

```python
def plot_decision_boundary(X, y, model):
    # Create a meshgrid
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))

    # Predict for grid points
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap='coolwarm')
    plt.title("Decision Tree Decision Boundary")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()

# Visualize
plot_decision_boundary(X, y,model)
```

## **Let us try with activation function as relu**

```python
model1 = Sequential()

model1.add(Dense(units = 1,activation = 'relu'))
model1.add(Dense(units = 1,activation = 'relu'))

model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model1.fit(X, y, epochs=300, batch_size=32)
```

```python
# Visualize
plot_decision_boundary(X, y,model1)
```

# **Let us try with activation function as Softmax**

```python
model11 = Sequential()

model11.add(Dense(units = 1,activation = 'softmax'))
model11.add(Dense(units = 1,activation = 'softmax'))

model11.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model11.fit(X, y, epochs=300, batch_size=32)
```

```python
# Visualize
plot_decision_boundary(X, y,model11)
```
